```python
!pip install yfinance statsmodels

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf

from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
import warnings
warnings.filterwarnings("ignore")
```

```python
# Download stock data
data = yf.download("AAPL", start="2020-01-01", end="2024-01-01")

# Use Closing Price
stock = data['Close']

# Plot
plt.figure(figsize=(10,5))
plt.plot(stock)
plt.title("AAPL Closing Price")
plt.show()
```

```python
result = adfuller(stock)
print("ADF Statistic:", result[0])
print("p-value:", result[1])
```

```python
stock_diff = stock.diff().dropna()

plt.figure(figsize=(12,5))

plt.subplot(121)
plot_acf(stock_diff, lags=20)

plt.subplot(122)
plot_pacf(stock_diff, lags=20)

plt.show()
```

```python
model_ar = ARIMA(stock, order=(2,1,0))   # AR(2)
model_ar_fit = model_ar.fit()

print(model_ar_fit.summary())
```

```python
model_ma = ARIMA(stock, order=(0,1,1))   # MA(1)
model_ma_fit = model_ma.fit()

print(model_ma_fit.summary())
```

```python
model_arma = ARIMA(stock, order=(2,1,1))  # ARMA(2,1)
model_arma_fit = model_arma.fit()

print(model_arma_fit.summary())
```

```python
print("AR Model AIC:", model_ar_fit.aic)
print("MA Model AIC:", model_ma_fit.aic)
print("ARMA Model AIC:", model_arma_fit.aic)

print("AR Model BIC:", model_ar_fit.bic)
print("MA Model BIC:", model_ma_fit.bic)
print("ARMA Model BIC:", model_arma_fit.bic)
```

```python
residuals = model_arma_fit.resid

plt.figure(figsize=(10,5))
plt.plot(residuals)
plt.title("Residuals")
plt.show()

plot_acf(residuals, lags=20)
plt.show()
```

```python
forecast = model_arma_fit.forecast(steps=30)

plt.figure(figsize=(10,5))
plt.plot(stock[-100:], label="Actual")
plt.plot(forecast, label="Forecast", color='red')
plt.legend()
plt.show()
```

```python
train = stock[:int(len(stock)*0.8)]
test = stock[int(len(stock)*0.8):]

model = ARIMA(train, order=(2,1,1))
fit = model.fit()

pred = fit.forecast(steps=len(test))

rmse = np.sqrt(mean_squared_error(test, pred))
print("RMSE:", rmse)
```
